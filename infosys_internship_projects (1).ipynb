```python
import os
import zipfile
import cv2
import matplotlib.pyplot as plt
import random
```

```python
zip_path = "/content/sample_data/archive.zip"
extract_path = "VOC2012"
if not os.path.exists(extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
print("Dataset extracted successfully!")
```

```python
image_dir = os.path.join(extract_path, "/content/VOC2012/voc_2012_segmentation_data/valid_images")
mask_dir = os.path.join(extract_path, "/content/VOC2012/voc_2012_segmentation_data/valid_labels")
```

```python
image_files = sorted(os.listdir(image_dir))
mask_files = sorted(os.listdir(mask_dir))

print("Total images:", len(image_files))
print("Total masks:", len(mask_files))
```

```python
def show_random_samples(n=5):
    plt.figure(figsize=(12, 8))
    for i in range(n):
        idx = random.randint(0, len(image_files) - 1)
        img_name = image_files[idx]
        img_path = os.path.join(image_dir, img_name)

        # Mask has same name but .png extension
        mask_name = img_name.replace(".jpg", ".png")
        mask_path = os.path.join(mask_dir, mask_name)

        # Load image and mask
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        mask = cv2.imread(mask_path)
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)

        # Show side by side
        plt.subplot(n, 2, i*2+1)
        plt.imshow(img)
        plt.title("Image")
        plt.axis("off")

        plt.subplot(n, 2, i*2+2)
        plt.imshow(mask)
        plt.title("Mask")
        plt.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
show_random_samples(n=5)
```
