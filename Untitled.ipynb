```python
import cv2, os
import matplotlib.pyplot as plt

images_path = "../images"
masks_path = "../output_masks"

for i, fname in enumerate(os.listdir(images_path)[:5]):  # show 5 samples
    img = cv2.imread(os.path.join(images_path, fname))
    mask = cv2.imread(os.path.join(masks_path, fname.replace('.jpg','_mask.png')), cv2.IMREAD_GRAYSCALE)

    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(mask, cmap="gray")
    plt.title("Binary Mask")
    plt.axis("off")

    plt.show()

```
